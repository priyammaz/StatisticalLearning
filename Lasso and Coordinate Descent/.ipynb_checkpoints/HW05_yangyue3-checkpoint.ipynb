**Answer to Question 1**

```python
# List of packages used
import numpy as np
import pandas as pd

from sklearn.preprocessing import scale
```

```python
np.random.seed(10)

n = 100
p = 200

# Generate data
V = np.ones((p,p),int)*0.3
np.fill_diagonal(V,1)
X_org = np.random.multivariate_normal(mean=np.zeros(p),cov=V,size=n)
true_b = np.zeros(p)
true_b[0:6] = [1,2,3,-3,-2,-1]
y_org = X_org@true_b.reshape(p,1)+np.random.standard_normal(n).reshape(n,1)

# Pre-scaling and centering
X = scale(X_org)*(n/(n-1))**0.5
y = scale(y_org)*(n/(n-1))**0.5
lambda_ = 0.3
```

1a: Implement soft threshold function and report results for testing cases. 

The first order condition with respect to the objective function is 

\begin{align}
0 = 
\begin{cases}
2(x-b)+\lambda \ \text{if} \ x \geq 0 \\
2(x-b)-\lambda \ \text{if} \ x < 0
\end{cases}
\end{align}

Therefore,

\begin{align}
x = 
\begin{cases}
\frac{2b-\lambda}{2} \ \text{if} \ b > \frac{\lambda}{2} \\
\frac{2b+\lambda}{2} \ \text{if} \ b < -\frac{\lambda}{2} \\
0 \ \text{otherwise}
\end{cases}
\end{align}



```python
def soft_th(b,lambda_):
  if b > lambda_/2:
    return (2*b-lambda_)/2
  elif b < -lambda_/2:
    return (2*b+lambda_)/2
  else:
    return 0

print('Output is {} when b is {}'.format(soft_th(1,0.3),1))
print('Output is {} when b is {}'.format(soft_th(-1,0.3),-1))
print('Output is {} when b is {}'.format(soft_th(-0.1,0.3),-0.1))
```

1b: Write a Lasso algorithm function. 

```python
# Lasso function (reuse the coordinate descent function from homework 4 and incorporate soft thresholding)
def my_lasso(X,y,lambda_,beta_init=np.zeros(p),max_iter=100,tol=1e-7):
  n = X.shape[0]
  p = X.shape[1]
  beta = beta_init
  betas = [beta_init.tolist()]
  iter = 0
  while iter < max_iter:
    for j in range(p):
      beta[j] = X[:,j].reshape(n,1).T@(y-np.delete(X,j,axis=1)@np.delete(beta.reshape(p,1),j,axis=0)) \
                /(X[:,j].reshape(n,1).T@X[:,j].reshape(n,1))
      beta[j] = soft_th(beta[j],lambda_) # soft thresholding
    betas.append(beta.tolist())
    adj = np.sum([abs(betas[-1][i]-betas[-2][i]) for i in range(p)])
    if adj <= tol:
      break
    iter += 1
  return betas,iter
```

```python
# Get the coefficients 
betas,num_iter = my_lasso(X,y,lambda_=0.3)
print('The first 8 coefficients are {} after {} iterations.'.format(betas[-1][:8],num_iter))
```

```python
# Compare with benchmark
glmnetfit_coefs_8 = [0.000000000,0.158233309,0.429648259,-0.519431173,-0.171467909,-0.006652507,0.000000000,0.000000000]
print('The L1 norm of the difference is {}.'.format(np.sum([abs(glmnetfit_coefs_8[i]-betas[-1][:8][i]) for i in range(8)])))
```

```python
# Read the glmnet output from R
glmnetfit_coefs = pd.read_csv('hw05_r_output1.csv')

discrepancies = []

for i in range(len(lambda_all)):
  discrepancy = np.sum([abs(glmnetfit_coefs.iloc[:,i][j].tolist()-betas_pw[i][j]) for j in range(p)])
  discrepancies.append(discrepancy)

results = pd.DataFrame({'log_lambda':np.log(lambda_all),'discrepancy':discrepancies})

plt.figure(figsize=(8,4))
sns.scatterplot(x='log_lambda',y='discrepancy',data=results)
plt.xlabel('Log($\lambda$)')
plt.ylabel('Discrepancies')
plt.title('Comparison with Benchmark')
plt.show()
```

```python
# Read the glmnet output from R
glmnetfit_coefs_org = pd.read_csv('hw05_r_output2.csv')

# Recover the coefficients with the unscaled_data
X_stds = [np.std(X_org[:,i]) for i in range(p)]
y_std = np.std(y_org)
betas_unscaled = []
for i in range(len(lambda_all)):
  betas_scaled = betas_pw[i]
  betas_recov = [betas_scaled[j]*y_std/X_stds[j] for j in range(p)]
  intercept_recov = [np.mean(y_org)-np.sum([np.mean(X_org[:,j])*betas_recov[j] for j in range(p)])]
  betas_recov = intercept_recov+betas_recov
  betas_unscaled.append(betas_recov)
```

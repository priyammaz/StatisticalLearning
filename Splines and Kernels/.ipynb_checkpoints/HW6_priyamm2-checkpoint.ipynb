### Part A: Implement Continuous Piecewise Linear Fitting

```python
knots = [0.2, 0.4, 0.6, 0.8]

def linear_splines(data, b0, b1, b2, b3, b4, b5):
    """
    Depending on the value of x, we have to calculate its value in a diffent spline. Our goal is to optimize
    out beta intercept and slopes for all our splines.
    """
    values = []
    for x in data:
        if x < 0.2:
            values.append(b0+b1*x)
        elif (x>=0.2)&(x<0.4):
            values.append(b0+b1*x+b2*(x-0.2))
        elif (x>=0.4)&(x<0.6):
            values.append(b0+b1*x+b2*(x-0.2)+b3*(x-0.4))
        elif (x>=0.6)&(x<0.8):
            values.append(b0+b1*x+b2*(x-0.2)+b3*(x-0.4)+b4*(x-0.6))
        elif (x>=0.8):
            values.append(b0+b1*x+b2*(x-0.2)+b3*(x-0.4)+b4*(x-0.6)+b5*(x-0.8))     
    return np.array(values)

parameters, _ = optimize.curve_fit(linear_splines,np.array(train['time']),np.array(train['ozone']))

plt.figure(figsize=(15,5))
plt.subplot(1,2,1)

sample_x = np.arange(0,1,0.01)
plot_values = linear_splines(sample_x, *parameters)

plt.subplot(1, 2, 1)
plt.scatter(train["time"], train["ozone"])
plt.plot(sample_x, plot_values, "red")
plt.title("Training Data: Time vs. Ozone")
plt.xlabel("Time")
plt.ylabel("Ozone")
plt.ylim(0, 40)  

plt.subplot(1, 2, 2)
plt.scatter(test["time"], test["ozone"], color="g")
plt.plot(sample_x, plot_values, "red")
plt.title("Testing Data: Time vs. Ozone")
plt.xlabel("Time")
plt.ylabel("Ozone")
plt.ylim(0, 40)  

plt.show()

linear_mse = mean_squared_error(np.array(test['ozone']),
                                linear_splines(np.array(test['time']),*parameters))
```

### Part B: Implement Quadratic Spline Fitting

Quadratic Spline Fitting is done by breaking out x variable into segments (based on input knots), and fitting a quadratic regression on each section with two conditions:

- Piecewise funcion is continuous for all values of x
- Piecewise function is differentiable for all values of x

Lets say we have some knot at value $a$.
Then our piecewise function would look like:

\begin{align}
y = 
\begin{cases}
b_0+b_{11}x+b_{12}x^2, \text{ if } x < a \\
b_1+b_{21}x+b_{22}x^2, \text{ if } x \geq a 
\end{cases}
\end{align}

**Condition 1**:

To satisfy our first condition, that the two quadratic equations be continuous at point $a$, they must equal to each other when $x=a$

$$b_0+b_{11}a+b_{12}a^2 = b_1+b_{21}a+b_{22}a^2$$

Instead of making a complex constraint like this for our optimizer, we can instead use a trick to incorporate the limitation into our basis functions

$$ \text{ if } x < a, y = b_0+b_{11}x+b_{12}x^2 $$
$$ \text{ elif } x \geq a, y = b_0+b_{11}a+b_{12}a^2+b_{21}x+b_{22}x^2 $$

Therefore, if we are greater than a, our intercept term of the quadratic equation becomes the value of the previous quadratic spline at the value of the knot


**Condition 2**:

To satisfy our second condition, the first derivative must be equal to each other between our two quadratic splines at the value of our knot


$$b_{11}+2b_{12}a = b_{21}+2b_{22}a$$

We already know the value of $b_{11}$ as it is the equation of the line of the previous spline that we are basing our next spline off of. We can then solve for $b_{21}$ 

$$b_{21} = b_{11}+2a(b_{12} - b_{22})$$

Therefore we only have to optimize on all coefficients of the initial spline and then only the coefficient of the polynomial term in the following splines. We will set the coefficient of the linear term in the following splines conditioned on maintaining our equal first derivative. Lastly we get the constant term of the following linear splines as the value of the previous spline calculated at that knot


```python
def quadratic_splines(data, b0, b11, b12, b22, b32, b42, b52):
    """
    We have to enforce that we are continuous and differentiable at all points
    """

    values = []
    for x in data:
        if x < 0.2:
            values.append(b0+b11*x+b12*x**2)
        elif (x>=0.2)&(x<0.4):
            values.append(b0+b11*x+b12*x**2+b22*(x-0.2)**2)
        elif (x>=0.4)&(x<0.6):
            values.append(b0+b11*x+b12*x**2+b22*(x-0.2)**2+b32*(x-0.4)**2)
        elif (x>=0.6)&(x<0.8):
            values.append(b0+b11*x+b12*x**2+b22*(x-0.2)**2+b32*(x-0.4)**2+b42*(x-0.6)**2)
        elif (x>=0.8):
            values.append(b0+b11*x+b12*x**2+b22*(x-0.2)**2+b32*(x-0.4)**2+b42*(x-0.6)**2+b52*(x-0.8)**2)     
    return np.array(values)

paramter_quadratic,_ = optimize.curve_fit(quadratic_splines,np.array(train['time']),np.array(train['ozone']))
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)

sample_x = np.arange(0,1,0.01)
plot_values = quadratic_splines(sample_x, *paramter_quadratic)

plt.subplot(1, 2, 1)
plt.scatter(train["time"], train["ozone"])
plt.plot(sample_x, plot_values, "red")
plt.title("Training Data: Time vs. Ozone")
plt.xlabel("Time")
plt.ylabel("Ozone")
plt.ylim(0, 40)  

plt.subplot(1, 2, 2)
plt.scatter(test["time"], test["ozone"], color="g")
plt.plot(sample_x, plot_values, "red")
plt.title("Testing Data: Time vs. Ozone")
plt.xlabel("Time")
plt.ylabel("Ozone")
plt.ylim(0, 40)  

plt.show()

quad_mse = mean_squared_error(np.array(test['ozone']),
                              quadratic_splines(np.array(test['time']),*paramter_quadratic))
```

**Part 1**: Check the difference between hat matricies where the difference is $max_{ij}|H_{ij}^{(1)} - H_{ij}^{(2)}|$

```python
def build_design_matrix(data):
    design = []
    for x in data:
        if x < 0.2:
            design.append([1,x,x**2,0,0,0,0])
        elif (x >= 0.2) & (x < 0.4):
            design.append([1,x,x**2,(x-0.2)**2,0,0,0])
        elif (x >= 0.4) & (x < 0.6):
            design.append([1,x,x**2,(x-0.2)**2,(x-0.4)**2,0,0])
        elif (x >= 0.6) & (x < 0.8):
            design.append([1,x,x**2,(x-0.2)**2,(x-0.4)**2,(x-0.6)**2,0])
        else:
            design.append([1,x,x**2,(x-0.2)**2,(x-0.4)**2,(x-0.6)**2,(x-0.8)**2])
    return np.array(design)

my_dmatrix = build_design_matrix(train.time)
my_hat_matrix = my_dmatrix@(np.linalg.inv(my_dmatrix.T@my_dmatrix))@my_dmatrix.T
bs_dmatrix = np.matrix(spline_dmatrix)
bs_hat_matrix = bs_dmatrix@(np.linalg.inv(bs_dmatrix.T@bs_dmatrix))@bs_dmatrix.T

err = np.max(np.abs(my_hat_matrix - bs_hat_matrix))
mse = np.abs(np.mean((my_hat_matrix - bs_hat_matrix)**2))
print("Max Absolute Difference:", round(err, 4))
print("Mean Squared Error Between Matricies", round(mse,5))
```

**Part 2**: Compare the conditional number $\frac{\sigma_{max}}{\sigma_{min}}$ of each design matrix where $\sigma_{max}$ and $\sigma_{min}$ denote the largest and smallest singular values

```python
S, V, D = np.linalg.svd(my_dmatrix)
my_condition_number = V[0]/V[-1]
print("Condition Number for My Design Matrix:", round(my_condition_number, 3))

S, V, D = np.linalg.svd(bs_dmatrix)
bs_condition_number = V[0]/V[-1]
print("Condition Number for BS Design Matrix:", round(bs_condition_number, 3))
```

## Question 2 [60 Points] Kernel Regression
### Part A:  One-dimensional kernel regression

**Gaussian Kernel**

```python
gauss_kernel = lambda x: (1/(2*math.pi)**0.5)*math.exp((-x**2)/2)

def gaussian_pred(x_range, x, y, lambda_value):
    predictions= []
    for xval in x_range:
        kernels = []
        for xi in x:
            kernels.append(gauss_kernel(abs(xi-xval)/lambda_value)/lambda_value)
        kernels = np.array(kernels)
        weight = kernels/np.sum(kernels)
        predictions.append(np.dot(weight, y))
    return predictions

def silverman(data):
    return 1.06*np.std(data)*len(data)**(-1/5)
x_axis = np.linspace(0,1,100)

silverman_lambda = silverman(train.time)
gauss_kernelout = gaussian_pred(x_axis, np.array(train.time), np.array(train.ozone), silverman_lambda)  
    

plt.figure(figsize=(15,5))
plt.subplot(1,2,1)

plt.subplot(1, 2, 1)
plt.scatter(train["time"], train["ozone"])
plt.plot(x_axis, gauss_kernelout, "red")
plt.title("Training Data: Time vs. Ozone")
plt.xlabel("Time")
plt.ylabel("Ozone")
plt.ylim(0, 40)  

plt.subplot(1, 2, 2)
plt.scatter(test["time"], test["ozone"], color="g")
plt.plot(x_axis, gauss_kernelout, "red")
plt.title("Testing Data: Time vs. Ozone")
plt.xlabel("Time")
plt.ylabel("Ozone")
plt.ylim(0, 40)  

plt.show()


gauss_test = gaussian_pred(np.array(test.time), np.array(train.time), np.array(train.ozone), silverman_lambda)  
gauss_err = mean_squared_error(gauss_test, test.ozone)
print("Mean Squared Error for Gaussian Kernel is: ", gauss_err)
```

**Epanechnikov Kernel**

```python
epan_kernel = lambda x: 0.75*(1-x**2) if abs(x) < 1 else 0

def epan_predict(x_range, x, y, lambda_value):
    predictions= []
    for xval in x_range:
        kernels = []
        for xi in x:
            kernels.append(epan_kernel(abs(xi-xval)/lambda_value)/lambda_value)
        kernels = np.array(kernels)
        weight = kernels/np.sum(kernels)
        predictions.append(np.dot(weight, y))
    return predictions

silverman_lambda = silverman(train.time)
epan_kernelout = epan_predict(x_axis, np.array(train.time), np.array(train.ozone), silverman_lambda) 

plt.figure(figsize=(15,5))
plt.subplot(1,2,1)

plt.subplot(1, 2, 1)
plt.scatter(train["time"], train["ozone"])
plt.plot(x_axis, epan_kernelout, "red")
plt.title("Training Data: Time vs. Ozone")
plt.xlabel("Time")
plt.ylabel("Ozone")
plt.ylim(0, 40)  

plt.subplot(1, 2, 2)
plt.scatter(test["time"], test["ozone"], color="g")
plt.plot(x_axis, epan_kernelout, "red")
plt.title("Testing Data: Time vs. Ozone")
plt.xlabel("Time")
plt.ylabel("Ozone")
plt.ylim(0, 40)  

plt.show()

epan_test = epan_predict(np.array(test.time), np.array(train.time), np.array(train.ozone), silverman_lambda)  
epan_err = mean_squared_error(epan_test, test.ozone)
print("Mean Squared Error for Gaussian Kernel is: ", epan_err)
```

**Tuning Epanechnikov Kernel**

```python
lambdas = np.linspace(0.02, 0.2, 10)

mse_values = []
for l in lambdas:
    epan_test = epan_predict(np.array(test.time), np.array(train.time), np.array(train.ozone), l)  
    mse_values.append(mean_squared_error(epan_test, test.ozone))
    
print(f"The Lambda that minimizes our MSE is {lambdas[np.argmin(mse_values)]}, with and MSE of {round(np.min(mse_values), 3)}")

epan_kernelout = epan_predict(x_axis, np.array(train.time), np.array(train.ozone),lambdas[np.argmin(mse_values)]) 

plt.figure(figsize=(15,5))
plt.subplot(1,2,1)

plt.subplot(1, 2, 1)
plt.scatter(train["time"], train["ozone"])
plt.plot(x_axis, epan_kernelout, "red")
plt.title("Training Data: Time vs. Ozone")
plt.xlabel("Time")
plt.ylabel("Ozone")
plt.ylim(0, 40)  

plt.subplot(1, 2, 2)
plt.scatter(test["time"], test["ozone"], color="g")
plt.plot(x_axis, epan_kernelout, "red")
plt.title("Testing Data: Time vs. Ozone")
plt.xlabel("Time")
plt.ylabel("Ozone")
plt.ylim(0, 40)  

plt.show()

```

### Part B: Two Dimensional Kernel Regression

```python
train_data = train[["time", "wind"]].values
train_ozone = train["ozone"].values
test_data = test[["time", "wind"]].values
test_ozone = test["ozone"].values
p = train_data.shape[1]
n = train_data.shape[0]

def multivar_silverman(data):
    return (4/(p+2))**(1/(p+4))*n**(-1/(p+4))*(np.std(data, axis=0))
lambda_values = multivar_silverman(train_data)
counter = 0
def multivar_gaussian_regression(grid, x, y, lambda_values):
    predictions = []
    for pair in grid: # Get every pair
        kernel_list = []
        for val in x:
            inner = np.sum(((val-pair)/lambda_values)**2)
            kernel_list.append(np.exp(-(inner)/2))
        kernel_list = np.array(kernel_list)
        weight = kernel_list / np.sum(kernel_list)
        predictions.append(weight@y)
        
    return predictions

multivar_pred = multivar_gaussian_regression(test_data, train_data, train_ozone, lambda_values)
multivar_mse = mean_squared_error(multivar_pred, test_ozone)

print("Error for Multivar Gaussian Kernel: ", multivar_mse)
```

### Part C: Variance of Multidimensional Kernel

```python
lambda_timeval = silverman(train.time)
lambda_windval = silverman(train.wind)
lambda_multivar = multivar_silverman(train_data)

print("Time Lambda:", lambda_timeval)
print("Wind Lambda:", lambda_windval)
print("Multivar Lambda:", lambda_multivar)
```

Lambda is a bandwidth (or smoothing) parameter. We can see that the silverman lambda for time and wind individually, but is higher when doing multivariate kernels. 

A higher lambda will smoothen the curve, but will lead to lower variance and high bias. On the other hand a low lambda will be a rougher curve but will give a higher variance with lower bias. 

Therefore we can see that the multivarite lambda will return a slightly higher bias and lower variance than what we would have gotten with the individual lambdas.

1a: Code a continuous priece-wise linear fitting from scratch.

```python
def pw_linear(x,b,k1,k2,k3,k4,k5):
  cond_list = [(x<0.2),
               (x>=0.2)&(x<0.4),
               (x>=0.4)&(x<0.6),
               (x>=0.6)&(x<0.8),
               (x>=0.8)]
  func_list = [lambda x:b+k1*x,
               lambda x:b+k1*x+k2*(x-0.2),
               lambda x:b+k1*x+k2*(x-0.2)+k3*(x-0.4),
               lambda x:b+k1*x+k2*(x-0.2)+k3*(x-0.4)+k4*(x-0.6),
               lambda x:b+k1*x+k2*(x-0.2)+k3*(x-0.4)+k4*(x-0.6)+k5*(x-0.8)]
  return np.piecewise(x,cond_list,func_list)
```

```python
para_linear,cov_linear = optimize.curve_fit(pw_linear,np.array(train['time']),np.array(train['ozone']))

x_axis = np.linspace(0,1,100)
plt.plot(np.array(train['time']),np.array(train['ozone']),"o",color='cyan',label='training_data')
plt.plot(np.array(test['time']),np.array(test['ozone']),"o",color='orange',label='testing_data')
plt.plot(x_axis,pw_linear(x_axis,*para_linear),color='red')
plt.xlabel('Time')
plt.ylabel('Ozone')
plt.title('Piece-wise Linear Fitting')
plt.legend()
plt.show()
```

```python
# Mean squared error
mse_pw_linear = mean_squared_error(np.array(test['ozone']),pw_linear(np.array(test['time']),*para_linear))
print('The testing data mean squared error from the piece-wise linear fitting is {}.'.format(mse_pw_linear)) 
```

1b: Code a continuous (up to the first derivative) quadratic spline fitting from scratch.

We first investigate the constraints we should impose to guarantee continuity up to first derivative.

We should have

\begin{align}
y = 
\begin{cases}
b_1+k_{11}x+k_{12}x^2 \quad &\text{if} \quad x < c \\
b_2+k_{21}x+k_{22}x^2 \quad &\text{if} \quad x \geq c
\end{cases}
\end{align}

We should have

\begin{align}
b1+k_{11}c+k_{12}c^2 &= b2+k_{21}c+k_{22}c^2 \\
k_{11}+2k_{12}c &= k_{21}+2k_{22}c 
\end{align}

We should incorporate these constraints into the piece-wise function we code.

```python
# Degrees of freedom is 7
def pw_quadratic(x,b1,k11,k12,k22,k32,k42,k52):
  k21 = k11+0.4*(k12-k22)
  k31 = k21+0.8*(k22-k32)
  k41 = k31+1.2*(k32-k42)
  k51 = k41+1.6*(k42-k52)

  cond_list = [(x<0.2),
               (x>=0.2)&(x<0.4),
               (x>=0.4)&(x<0.6),
               (x>=0.6)&(x<0.8),
               (x>=0.8)]

  func_list = [lambda x:b1+k11*x+k12*x**2,
               lambda x:b1+k11*x+k12*x**2+k21*(x-0.2)+k22*(x-0.2)**2,
               lambda x:b1+k11*x+k12*x**2+k21*(x-0.2)+k22*(x-0.2)**2+k31*(x-0.4)+k32*(x-0.4)**2,
               lambda x:b1+k11*x+k12*x**2+k21*(x-0.2)+k22*(x-0.2)**2+k31*(x-0.4)+k32*(x-0.4)**2+k41*(x-0.6)+k42*(x-0.6)**2,
               lambda x:b1+k11*x+k12*x**2+k21*(x-0.2)+k22*(x-0.2)**2+k31*(x-0.4)+k32*(x-0.4)**2+k41*(x-0.6)+k42*(x-0.6)**2+k51*(x-0.8)+k52*(x-0.8)**2]
  return np.piecewise(x,cond_list,func_list)
```

```python
para_quadratic,cov_quadratic = optimize.curve_fit(pw_quadratic,np.array(train['time']),np.array(train['ozone']))

x_axis = np.linspace(0,1,100)
plt.plot(np.array(train['time']),np.array(train['ozone']),"o",color='cyan',label='training_data')
plt.plot(np.array(test['time']),np.array(test['ozone']),"o",color='orange',label='testing_data')
plt.plot(x_axis,pw_quadratic(x_axis,*para_quadratic),color='red')
plt.xlabel('Time')
plt.ylabel('Ozone')
plt.title('Piece-wise Quadratic Fitting')
plt.legend()
plt.show()
```

```python
# Mean squared error
mse_pw_quadratic = mean_squared_error(np.array(test['ozone']),pw_quadratic(np.array(test['time']),*para_quadratic))
print('The testing data mean squared error from the piece-wise quadratic fitting is {}.'.format(mse_pw_quadratic)) 
```

The hat matrix is

\begin{align}
H = X(X^{'}X)^{-1}X^{'}
\end{align}

where $X$ is the design matrix for the linear regression.

```python
def design_quad(x):
  if x < 0.2:
    return [1,x,x**2,0,0,0,0,0,0,0,0]
  elif (x >= 0.2) & (x < 0.4):
    return [1,x,x**2,(x-0.2),(x-0.2)**2,0,0,0,0,0,0]
  elif (x >= 0.4) & (x < 0.6):
    return [1,x,x**2,(x-0.2),(x-0.2)**2,(x-0.4),(x-0.4)**2,0,0,0,0]
  elif (x >= 0.6) & (x < 0.8):
    return [1,x,x**2,(x-0.2),(x-0.2)**2,(x-0.4),(x-0.4)**2,(x-0.6),(x-0.6)**2,0,0]
  else:
    return [1,x,x**2,(x-0.2),(x-0.2)**2,(x-0.4),(x-0.4)**2,(x-0.6),(x-0.6)**2,(x-0.8),(x-0.8)**2]
```

```python
# Hat matrix for piece-wise quadratic
def design_quad(x):
  if x < 0.2:
    return [1,x,x**2,0,0,0,0,0,0,0,0]
  elif (x >= 0.2) & (x < 0.4):
    return [1,x,x**2,(x-0.2),(x-0.2)**2,0,0,0,0,0,0]
  elif (x >= 0.4) & (x < 0.6):
    return [1,x,x**2,(x-0.2),(x-0.2)**2,(x-0.4),(x-0.4)**2,0,0,0,0]
  elif (x >= 0.6) & (x < 0.8):
    return [1,x,x**2,(x-0.2),(x-0.2)**2,(x-0.4),(x-0.4)**2,(x-0.6),(x-0.6)**2,0,0]
  else:
    return [1,x,x**2,(x-0.2),(x-0.2)**2,(x-0.4),(x-0.4)**2,(x-0.6),(x-0.6)**2,(x-0.8),(x-0.8)**2]

X_quad = np.array([design_quad(i) for i in np.array(train.time)])
hat_quad = X_quad@(np.linalg.inv(X_quad.T@X_quad))@X_quad.T

# Hat matrix for the b-spline 
X_bs = np.matrix(basis_bspline)
hat_bspline = X_bs@(np.linalg.inv(X_bs.T@X_bs))@X_bs.T

# Difference
diff = np.max(np.abs(hat_quad-hat_bspline))
print('The difference between the two design matrices is {}.'.format(diff))

# Singluar values
sing_quad = (np.linalg.eig(X_quad.T@X_quad))[0]**0.5
sing_bs = (np.linalg.eig(X_bs.T@X_bs))[0]**0.5
cond_quad = np.max(sing_quad)/np.min(sing_quad)
cond_bs = np.max(sing_bs)/np.min(sing_bs)
print('The conditional number for piece-wise quadratic fitting and b-spline fitting are {} and {}, respectively.'.format(cond_quad,cond_bs))
```

B-spline's estimates are more accurate than the piece-wise quadratic estamtes since it measures the bound on how inaccurate the solution (estimates) will be after approximation.

```python
# List of packages used
import math
import itertools

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

2a: Code Gaussian and Epanechnikov kernel regression model. 

```python
# Gaussian kernel
def gaussian_kernel(z):
  return (1/math.sqrt(2*math.pi))*math.exp(-z**2/2)

def gaussian_predict(x0,x,y,lambda_):
  n = len(x)
  kernels = [gaussian_kernel(np.abs(xi-x0)/lambda_)/lambda_ for xi in x]
  weights = [kernel/np.sum(kernels) for kernel in kernels]
  return weights@y
```

```python
# Silverman lambda
silverman_lambda_ = 1.06*np.std(np.array(train.time))*len(np.array(train.time))**(-0.2)
x_axis = np.linspace(0,1,100)

# Simulated y values with Gaussian kernel
simulated_y_gauss_kernel = np.array([gaussian_predict(x0,np.array(train.time),np.array(train.ozone),silverman_lambda_) for x0 in x_axis])

plt.plot(np.array(train['time']),np.array(train['ozone']),"o",color='cyan',label='training_data')
plt.plot(np.array(test['time']),np.array(test['ozone']),"o",color='orange',label='testing_data')
plt.plot(x_axis,simulated_y_gauss_kernel,color='red')
plt.xlabel('Time')
plt.ylabel('Ozone')
plt.title('Gaussian Kernel Fitting (Silverman Lambda)')
plt.legend()
plt.show()
```

```python
# Mean squared error
y_pred_gauss_kernel = np.array([gaussian_predict(x0,np.array(train.time),np.array(train.ozone),silverman_lambda_) for x0 in np.array(test.time)])
mse_gaussian_kernel = mean_squared_error(np.array(test['ozone']),y_pred_gauss_kernel)
print('The testing data mean squared error from the Gaussian kernel fitting with Silverman lambda is {}.'.format(mse_gaussian_kernel)) 
```

```python
# Epanechnikov kernel
def epanechnikov_kernel(z):
  return 0.75*(1-z**2) if np.abs(z) <= 1 else 0

def epanechnikov_predict(x0,x,y,lambda_):
  n = len(x)
  kernels = [epanechnikov_kernel(np.abs(xi-x0)/lambda_)/lambda_ for xi in x]
  weights = [kernel/np.sum(kernels) for kernel in kernels]
  return weights@y
```

```python
# Silverman lambda
silverman_lambda_ = 1.06*np.std(np.array(train.time))*len(np.array(train.time))**(-0.2)
x_axis = np.linspace(0,1,100)

# Simulated y values with Epanechnikov kernel
simulated_y_epa_kernel = np.array([epanechnikov_predict(x0,np.array(train.time),np.array(train.ozone),silverman_lambda_) for x0 in x_axis])

plt.plot(np.array(train['time']),np.array(train['ozone']),"o",color='cyan',label='training_data')
plt.plot(np.array(test['time']),np.array(test['ozone']),"o",color='orange',label='testing_data')
plt.plot(x_axis,simulated_y_epa_kernel,color='red')
plt.xlabel('Time')
plt.ylabel('Ozone')
plt.title('Epanechnikov Kernel Fitting (Silverman Lambda)')
plt.legend()
plt.show()
```

```python
# Mean squared error
y_pred_epa_kernel = np.array([epanechnikov_predict(x0,np.array(train.time),np.array(train.ozone),silverman_lambda_) for x0 in np.array(test.time)])
mse_epa_kernel = mean_squared_error(np.array(test['ozone']),y_pred_epa_kernel)
print('The testing data mean squared error from the Epanechinov kernel fitting with Silverman lambda is {}.'.format(mse_epa_kernel)) 
```

```python
# Tuning lambda value for Epanechnikov kernel (Silverman lambda is 0.1013)
lambda_list = np.linspace(0.05,0.2,10)

mse_search = []
for lambda_ in lambda_list:
  y_pred = np.array([epanechnikov_predict(x0,np.array(train.time),np.array(train.ozone),lambda_) for x0 in np.array(test.time)])
  mse_search.append(mean_squared_error(np.array(test['ozone']),y_pred))

mse_min_lambda_ = lambda_list[np.argmin(mse_search)]
print('The testing mean squared error minimizing lambda is {} with mean squared error {}.'.format(mse_min_lambda_,np.min(mse_search)))
```

```python
# Simulated y values with Epanechnikov kernel
simulated_y_epa_kernel = np.array([epanechnikov_predict(x0,np.array(train.time),np.array(train.ozone),mse_min_lambda_) for x0 in x_axis])

plt.plot(np.array(train['time']),np.array(train['ozone']),"o",color='cyan',label='training_data')
plt.plot(np.array(test['time']),np.array(test['ozone']),"o",color='orange',label='testing_data')
plt.plot(x_axis,simulated_y_epa_kernel,color='red')
plt.xlabel('Time')
plt.ylabel('Ozone')
plt.title('Epanechnikov Kernel Fitting (Tuned Lambda)')
plt.legend()
plt.show()
```

2b: Implement two-dimensional kernel.

```python
# Two-dimensional Gaussian kernel
def gaussian_kernel_multi(z):
  return np.exp(-0.5*z**2)

# Note that new_data should be an array of [x,y]
def gaussian_predict_multi(new_data,x,y,lambda_):
  estimates = []
  for d in new_data:
    kernels = np.array([gaussian_kernel_multi(np.linalg.norm(np.divide((xi-d),lambda_))) for xi in x])
    weights = np.array([kernel/np.sum(kernels) for kernel in kernels])
    estimates.append(weights@y)
  return estimates
```

```python
# Silverman lambda
p = 2
silverman_lambda_ = (4/(p+2))**(1/(p+4))*train.shape[0]**(-1/(p+4))*np.array([np.std(train.time),np.std(train.wind)])
time_axis = np.linspace(0,1,100)
wind_axis = np.linspace(0,11,12)
new_data = np.array(list(itertools.product(time_axis,wind_axis)))
time,wind = np.meshgrid(time_axis,wind_axis)

# Simulated y values with Epanechnikov kernel
simulated_y_gaussian_multi = gaussian_predict_multi(new_data,np.array(train[['time','wind']]),np.array(train.ozone),silverman_lambda_)
simulated_y_gaussian_multi = np.array(simulated_y_gaussian_multi).reshape(time.shape[0],time.shape[1])

ax = plt.axes(projection='3d')
ax.scatter3D(np.array(train['time']),np.array(train['wind']),np.array(train['ozone']),"o",color='cyan',label='training_data')
ax.scatter3D(np.array(test['time']),np.array(test['wind']),np.array(test['ozone']),"o",color='orange',label='testing_data')
ax.plot_surface(time,wind,simulated_y_gaussian_multi,cmap='viridis',edgecolor='none',antialiased=True,alpha=0.5)
ax.set_xlabel('Time')
ax.set_ylabel('Wind')
ax.set_zlabel('Ozone')
plt.title('Multi-variate Gaussian Kernel Fitting (Silverman Lambda)')
plt.legend(bbox_to_anchor=(1.5,1))
plt.show()
```

```python
# Mean squared error
y_pred_multi_gauss_kernel = np.array(gaussian_predict_multi(np.array(test[['time','wind']]),np.array(train[['time','wind']]),np.array(train.ozone),silverman_lambda_))
mse_multi_gauss_kernel = mean_squared_error(np.array(test['ozone']),y_pred_multi_gauss_kernel)
print('The testing data mean squared error from the Multi-variate Gaussian kernel fitting with Silverman lambda is {}.'.format(mse_multi_gauss_kernel)) 
```

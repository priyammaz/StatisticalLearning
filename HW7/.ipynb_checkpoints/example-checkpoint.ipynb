**Answer to Question 1**

```python
# List of packages used
import math

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

1a: Write a local linear regression function. 

```python
# Gaussian kernel
def gaussian_kernel(z):
  return np.exp(-0.5*z**2)

# Weight matrix
def weight_matrix(point,X_train,lambda_):
  m = X_train.shape[0]
  w = np.mat(np.identity(m)) # initialize weight matrix
  for i in range(m):
    x_i = X_train[i]
    kernel = gaussian_kernel(np.linalg.norm(np.divide((x_i-point),lambda_)))/lambda_
    w[i,i] = kernel
  return w

# Point prediction function
def point_pred(X_train,y_train,test_point,lambda_):
  bias = np.ones((m,1))
  X_train_ext = np.hstack((bias,X_train))
  test_point_ext = np.mat([1,test_point])
  w = weight_matrix(test_point_ext,X_train_ext,lambda_)
  beta = np.linalg.inv(X_train_ext.T@w@X_train_ext)@(X_train_ext.T@w@y_train)
  pred = test_point_ext@beta
  return beta,pred

# Local linear function
def my_local_linear(X_train,y_train,X_test,lambda_):
  preds = []
  for point in np.ravel(X_test):
    beta,pred = point_pred(X_train,y_train,point,lambda_)
    preds.append(pred)
  return np.ravel(preds)
```

1b: Make predictions and report results.

```python
# Preprocess the training data
df_train = pd.read_csv('hw7_Q1_train.csv')
m = df_train.shape[0]
X_train = np.array(df_train.x).reshape(m,1)
y_train = np.array(df_train.y).reshape(m,1)

# Generate the testing data
X_test = 2*math.pi*np.linspace(0,1,101)
y_test = np.array([math.sin(val) for val in X_test])
X_test = X_test.reshape(101,1)
y_test = y_test.reshape(101,1)

bandwidths = np.linspace(0.05,0.6,12)
```

```python
# Mean squared errors
def mse(y_test,y_pred):
  return np.mean((y_test-y_pred)**2)

mses = []
for b in bandwidths:
  y_pred = my_local_linear(X_train,y_train,X_test,b)
  mses.append(mse(np.ravel(y_test),y_pred))
```

```python
plt.plot(bandwidths,mses)
plt.title('Mean Squared Errors vs. Bandwidths')
plt.xlabel('Bandwidths')
plt.ylabel('MSEs')
plt.show()
```

Roughly speaking, the plot does have a "U" shape.

```python
best_mse = np.min(mses)
best_lambda_ = bandwidths[np.argmin(mses)]
print('The best MSE is {} with lambda value {}.'.format(best_mse,best_lambda_))
```

```python
lambdas_ = [0.05,0.25,0.5]

f,a = plt.subplots(1,3,figsize=(10,4))
for i in range(3):
  y_pred = my_local_linear(X_train,y_train,X_test,lambdas_[i])
  a[i].scatter(X_test,y_pred,s=1,c='red',label='Predicted')
  a[i].scatter(X_train,y_train,s=1,c='black',label='Training',alpha=0.5)
  a[i].plot(X_test,y_test,label='Truth')
  a[i].set_title('lambda='+str(lambdas_[i]))
  a[i].set_xlabel('X_test')
  a[i].set_ylabel('y')
  a[i].legend()

plt.tight_layout()
plt.show()
```

As $\lambda$ increases, the prediction variance decreases (the fitted curve becomes smoother). 

**Answer to Question 2**

```python
# List of packages used
import numpy as np
import pandas as pd
```

```python
# Process data
train = pd.read_csv('digits_train.csv')
test = pd.read_csv('digits_test.csv')

X_train = train.iloc[:,1:]
X_test = test.iloc[:,1:]
y_train = train.iloc[:,0]
y_test = test.iloc[:,0]

num_classes = y_train.nunique()
```

2a: Estimate the mean, covariance matrix of each class and pooled covariance matrix.

```python
# Mean, class covariance, pooled covariance
means = np.zeros((10,256))
cov_class = np.array([np.zeros((256,256)) for i in range(10)])
cov_pool = np.zeros((256,256))
count = [sum(map(lambda x:x==i,y_train)) for i in range(10)]

for i in range(10):
  print(i)
  cov_array = []
  for j in range(X_train.shape[0]):
    if y_train[j] == i:
      means[i] = means[i]+np.array(X_train)[j]
      cov_array.append(np.array(X_train)[j])
  cov_class[i] = np.cov(np.array(cov_array).T)*(count[i]/(count[i]-1))
  cov_pool = cov_pool+ np.cov(np.array(cov_array).T)*(count[i]/(y_train.shape[0]-num_classes))
  means[i] = means[i]/count[i]


```

```python
cov_pool
```

2b: Write LDA function from scratch.

```python
def my_lda(X_test,y_test,means,cov_pool):
  inv_cov_pool = np.linalg.inv(cov_pool)
  confusion_matrix = np.zeros((num_classes,num_classes),dtype='int')
  failure = []
  predictions = []
  for i in range(10):
    for j in range(y_test.shape[0]):
      if y_test[j] == i:
        p = np.zeros(10)
        for k in range(10):
          p[k] = means[k].T@inv_cov_pool@np.array(X_test)[j]-(means[k].T@inv_cov_pool@means[k])/2+np.log(count[i]/y_train.shape[0])
        pred = p.argmax()
        if pred != y_test[j]:
          failure.append((j,pred))
        confusion_matrix[pred][y_test[j]] += 1
        predictions.append(pred)
  return predictions,confusion_matrix,failure
```

2c: Fit the data and make predictions. 

```python
# First 5 entries of the coefficient vector and bias for digit 0
w0 = np.linalg.inv(cov_pool)@(means[0].reshape(-1,1))
print('The first 5 entries of the coefficient vector are {}'.format(np.ravel(w0)[:5]))

b0 = -(means[0].T@np.linalg.inv(cov_pool)@means[0])/2+np.log(count[0]/y_train.shape[0])
print('The bias is {}'.format(np.ravel(b0)))
```

```python
# Predictions
predictions,confusion_matrix,failure = my_lda(X_test,y_test,means,cov_pool)
```

```python
# Confusion matrix
df_confusion = pd.DataFrame(confusion_matrix,columns=[i for i in range(10)])
df_confusion
```

```python
# Misclassification rate
miss_class = {}
for i in range(10):
  miss_class['digit '+str(i)] = np.round(1-df_confusion.iloc[i,i]/df_confusion.iloc[:,i].sum(),3)

print('The misclassification rate of each digit is \n{}.'.format(pd.Series(miss_class)))
print('The overall misclassification rate is {}.'.format(np.round(1-np.trace(confusion_matrix)/y_test.shape[0],3)))
```

```python
# List of packages used
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

3a: Write regularized QDA function from scratch. Note that we have estimated the per class covariance in question 2.

```python
def my_rqda(X_test,y_test,means,cov_class,cov_pool,alpha):
  confusion_matrix = np.zeros((num_classes,num_classes),dtype='int')
  failure = []
  predictions = []
  cov_class_new = [alpha*cov_class[i]+(1-alpha)*cov_pool for i in range(10)]
  inv_cov_class_new = [np.linalg.inv(cov_class_new[i]+0.00001*np.eye(256)) for i in range(10)]
  for i in range(10):
    for j in range(y_test.shape[0]):
      if y_test[j] == i:
        p = np.zeros(10)
        for k in range(10):
          cov_k = cov_class_new[k]
          inv_cov_k = inv_cov_class_new[k]
          p[k] = -np.log(np.linalg.det(cov_k)+0.0001)/2-((np.array(X_test)[j]-means[k]).T@inv_cov_k@(np.array(X_test)[j]-means[k]))/2+np.log(count[i]/y_train.shape[0])
        pred = p.argmax()
        if pred != y_test[j]:
          failure.append((j,pred))
        confusion_matrix[pred][y_test[j]] += 1
        predictions.append(pred)
  return predictions,confusion_matrix,failure
```

3b: Compare different $\alpha$ values.

```python
alphas = [0,0.05,0.1,0.15,0.2,0.25,0.3]
miss_class_rates = []

for alpha in alphas:
  predictions,confusion_matrix,failure = my_rqda(X_test,y_test,means,cov_class,cov_pool,alpha)
  rate = np.round(1-np.trace(confusion_matrix)/y_test.shape[0],3)
  miss_class_rates.append(rate)
  print('alpha {}: rate {}'.format(alpha,rate))
```

```python
plt.plot(alphas,miss_class_rates)
plt.title('Misclassification Rates vs. Regularization Weights')
plt.xlabel('Alphas')
plt.ylabel('Misclassification Rates')
plt.show()
```

```python
print('The lowest testing error is {} with alpha value {}.'.format(np.min(miss_class_rates),alphas[np.argmin(miss_class_rates)]))
```

## Question 1: Lasso Solution for Fixed $\lambda$

```python
### Import Packages ###
import numpy as np
import pandas as pd
from sklearn.preprocessing import scale
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
### Generate Data ###
np.random.seed(10)
n = 100
p = 200

# Generate data
V = np.ones((p,p))*0.3
np.fill_diagonal(V,1)

X_org = np.random.multivariate_normal(mean=np.zeros(p),cov=V,size=n)
true_b = np.hstack([np.array([1,2,3,-3,-2,-1]), np.zeros(p-6)])
y_org = X_org@true_b.reshape(p,1)+np.random.standard_normal(n).reshape(n,1)
# Pre-scaling and centering
X = scale(X_org, axis=0)*(n/(n-1))**0.5
y = scale(y_org, axis=0)*(n/(n-1))**0.5
lambda_value = 0.3

print("X Shape:", X.shape, "Y Shape", y.shape)
```

**Part A**

Take the first derivative of the equation, and we get two equations depending on the value of x


\begin{equation}
  0 =\begin{cases}
    2(x-b) + \lambda, & \text{if $x\geq0$}.\\
    2(x-b) - \lambda, & \text{if x < 0}.
  \end{cases}
\end{equation}

\begin{equation}
  x =\begin{cases}
    \frac{2b-\lambda}{2}, & \text{if $b > \frac{\lambda}{2}$}.\\
    \frac{2b+\lambda}{2}, & \text{if $b < \frac{-\lambda}{2}$}.\\
    0, &  \text{otherwise}
  \end{cases}
\end{equation}



```python
def soft_th(b,lambda_value):
    if b > lambda_value/2:
        return (2*b-lambda_value)/2
    elif b < -lambda_value/2:
        return (2*b+lambda_value)/2
    else:
        return 0
    
print('When b is {}: {}'.format(1, soft_th(1,0.3)))
print('When b is {}: {}'.format(-1, soft_th(-1,0.3)))
print('When b is {}: {}'.format(-0.1, soft_th(-0.1,0.3),))
```

**Part B**

```python
def my_lasso(X,y,lambda_value, beta_init=None, max_iter=100,tol=1e-7):
    n, p = X.shape
    if beta_init is not None:
        beta = np.array(beta_init)
    else:
        beta = np.zeros(p)
    betas = [beta.tolist()]
    counter = 0
    while True:
        if counter >= max_iter:
            break
        for j in range(p):
            x_j =  X[:,j].reshape(n,1)
            x_wo_j = np.delete(X,j,axis=1)
            b_wo_j = np.delete(beta.reshape(p,1),j,axis=0)
            beta[j] = x_j.T@(y-x_wo_j@b_wo_j) / (x_j.T@x_j)
            beta[j] = soft_th(beta[j],lambda_value)
        betas.append(beta.tolist())
        change = sum(abs(betas[-1][i] - betas[-2][i]) for i in range(p))
        if change < tol:
            break
        counter += 1
    return betas, counter
            
betas, counter = my_lasso(X, y, 0.3)  

print("First 8 Coefficients: {}".format(betas[-1][0:8]))
print("Number of Iterations: {}".format(counter))


# Compare to GLMTest
glm_coeff = [0.000000000,0.158233309,0.429648259,-0.519431173,
            -0.171467909,-0.006652507,0.000000000,0.000000000]
print('The L1 norm {}.'.format(np.sum([abs(glm_coeff[i]-betas[-1][:8][i]) for i in range(8)])))
```

**Part A**

```python
def myLasso_pw(X, y, lambda_all, max_iter=100,tol=1e-7):
    beta = np.zeros(p)
    betas = [beta.tolist()]
    for idx, lambda_value in tqdm(enumerate(lambda_all), total=len(lambda_all)):
        beta_l, _ = my_lasso(X, y,lambda_value, beta_init=betas[idx])
        betas.append(beta_l[-1])
    return betas

betas = myLasso_pw(X,y, lambda_all)[1:]
```
